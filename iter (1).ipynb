```python
import random
import itertools
def randomPerm(n):
    v=[]
    for i in range(n):
        v.append(i+1)
    for i in range(len(v)-1):
        j = random.randint(i, len(v)-1)
        aux = v[i]
        v[i] = v[j]
        v[j] = aux
    return v

def isortSteps(a):
    compi = 0
    cambios = 0
    precicl = 2
    v = []
    for i in range(len(a)):
        v.append(a[i])        
    steps = 0
    for i in range(1,len(v)):
        x = v[i]
        j = i-1
        while (j > -1) and (v[j] > x):
            precicl += 2
            if v[j] > x:
                compi +=  1
            v[j+1] = v[j]
            cambios += 1
            j = j -1
            steps = steps + 3
        steps = steps + 1
        v[j+1] = x
        cambios += 1
        steps = steps + 4
    steps = steps + 1
    print 'pasos: ' + str(steps)
    print 'comparaciones: ' + str(compi)
    print 'intercambios: ' + str(cambios)
    print 'preguntas de ciclo: ' + str(precicl)
    return steps


n = 6

sum = 0.0
min = n**3
max = 0.0

x = randomPerm(n)
count = 0
for i in itertools.permutations(x):
    print i
    t = isortSteps(i)
    count = count + 1
    if t < min :
       min = t
    if t > max :
       max = t
    sum = sum + t 

print 'n: ' + str(n)
print 'Permutations: ' + str(count)
print 'Theoretical best time, ' + str(5*n - 4) 
print 'Theoretical worst time,' + str((3.0/2.0)*n*2 + (7.0/2.0)*n - 4)
print 'Theoretical average time,' + str((3.0/4.0)*n*2 + (17.0/4.0)*n - 4)
print 'Experimenal best time, ' + str(min)
print 'Experimenal worst time,' + str(max)
print 'Experimenal average time, ' + str(sum/count)
```
